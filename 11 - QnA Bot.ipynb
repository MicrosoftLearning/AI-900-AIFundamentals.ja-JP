# 対話型 AI

インスタント メッセージング、ソーシャル メディア、電子メールなどのオンライン テクノロジを使って、どれくらい頻繁に他の人とコミュニケーションしているか考えてください。多くの人々は、これらの連絡手段を頼りにしています。職場でわからないことがあっても、モバイル デバイスで使えるチャット メッセージで同僚と連絡をとれるので、常にみんなとつながっています。

![会話している人間とロボット](./images/conversational_ai.jpg)

ボットは、これらの連絡手段を使用してコミュニケーションする AI エージェントであり、ソフトウェア サービスとの自然な会話を可能にします。

## QnA Maker ナレッジ ベースを作成する

カスタマー サポートの場面では、一般に Web サイト上でチャット画面、電子メール、音声などのインターフェイスを使い、よくある質問を理解して回答できるボットが作成されます。ボット インターフェイスの基礎にあるのは質問とそれへの適切な回答の知識ベースであり、ボットはこれを検索して適切に応答することができます。

QnA Maker サービスは Azure Cognitive Service の 1 つであり、作成した質問と回答のペアや、既存のドキュメントや Web ページから、ナレッジ ベースをすばやく作成できます。ナレッジ ベースができれば、自然言語を処理する組み込みの機能を使用して、質問を解釈し、適切な回答を見つけることができます。

1. ブラウザーで新しいタブを開き、https://qnamaker.ai の QnAMaker ポータルにアクセスします。Azure サブスクリプションに関連付けられている Microsoft アカウントを使用してサインインします。
2. QnA Maker ポータルで、「**Create a knowledge base (ナレッジ ベースの作成)**」 を選択します。
3. QnA サービス リソースを作成していない場合は、「**Create a QnA service (QnA サービスの作成)**」 を選択します。新しいタブで Azure portal が開き、QnA Maker サービスを Azure サブスクリプションで作成できます。次の設定を使用します。
   
   - **サブスクリプション**: *使用する Azure サブスクリプション*
   - **リソース グループ**: *既存のリソース グループを選択するか、新しいリソース グループを作成します*
   - **名前**: *使用する QnA リソースの一意の名前*
   - **価格レベル**: F0
   - **Azure Search の価格レベル**: F
   - **Azure Search の場所**: *利用可能な任意の場所*
   - **アプリ名**: _**名前**と同じ (自動的に ".azurewebsites.net" が付加されます)_
   - **Web サイトの場所**: _**Azure Search の場所**と同じ_
   - **アプリ インサイト**: 無効

> **注**: Free レベルの **QnA Maker** リソース、または **Azure Search** リソースを既にプロビジョニングしている場合、クォータによって新しいリソースを作成できないことがあります。その場合、**F0** や **F** 以外の価格レベルを選択してください。

4. Azure portal で QnA サービスと関連リソースのデプロイが完了するのを待ちます。
5. QnA Maker のポータル タブに戻り、「**Step 2**」 セクションで 「**Refreesh (更新)**」 をクリックして、使用できる QnA サービス リソースの一覧を更新します。
6. 次のオプションを選択して、QnA サービスを KB に接続します。
  - **Microsoft Azure Directory ID**: *お使いの Azure サブスクリプションの Directory ID*
  - **Azure サブスクリプション名**: *使用する Azure サブスクリプション*
  - **Azure QnA サービス**: *前の手順で作成した QnA サービス リソース*
  - **言語**: 英語

### (!)確認
アクションを実行する権限がロールにないことを示すメッセージが表示された場合は、QnA Maker ポータルのページをブラウザーで更新してください。

7. 「**Step 3**」 セクションで、名前「**Margie's Travel KB**」を入力します。
8. 「**Step 4**」 セクションの 「**URL**」 ボックスに、*https://github.com/MicrosoftLearning/mslearn-ai900/raw/main/data/qna_bot/margies_faq.docx* と入力し、「**+ Add URL (+ URL を追加)**」 をクリックします。次に、「**Chit-chat (チットチャット)**」 で 「**Professional (プロフェッショナル)**」 を選択します。
9. 「**Step 5**」 セクションの 「**Create your KB (KB の作成)**」 をクリックします。
10. ナレッジ ベースが作成されるまで 1 分ほど待ちます。そのあと、FAQ ドキュメントからインポートされた質問と回答、および事前定義された仕事上のチットチャットの応答を確認します。


## ナレッジ ベースを編集する

ナレッジ ベースは、FAQ ドキュメントの内容と、事前定義された応答を基にしています。カスタマイズした質問と回答のペアを追加して、内容を補足することもできます。

1. 「**+ Add QnA pair (+ QnA ペアの追加)**」 をクリックします。
2. 「**Question (質問)**」 ボックスに`Hello`と入力します。次に 「**+ Add alternative phrasing (+ 代わりのフレーズを追加)**」 をクリックして`Hi`と入力します。
3. 「**Answer (回答)**」 ボックスに`Hello`と入力します。

## ナレッジ ベースのトレーニングとテスト

これでナレッジ ベースの準備ができました。次に QnA Maker ポータルでテストを行います。

1. ページの右上にある 「**Save and train (保存してトレーニング)**」 をクリックして、ナレッジ ベースをトレーニングします。画面を最大に広げないとボタンが表示されない場合があります。
2. トレーニングが完了したら、「**&larr; Test (テスト)**」 をクリックしてテスト画面を開きます。
3. テスト画面の下部に、「*Hi*」というメッセージを入力します。応答「**Hello**」が返ります。
4. テスト ペインの下部に、「*フライトを予約したい* (I want to book a flight)」というメッセージを入力します。FAQ から適切な応答が返るはずです。
5. ナレッジ ベースのテストが終了したら、「**&rarr; Test (テスト)**」 をクリックしてテスト画面を閉じます。


## ナレッジ ベース用のボットを作成する

クライアント アプリケーションは、何らかのユーザー インターフェイスを介して質問に回答するとき、ナレッジ ベースが提供するバックエンド サービスを利用できます。一般的に、クライアント アプリケーションはボットです。ボットがナレッジ ベースを利用できるようにするには、ナレッジ ベースを HTTP 経由でアクセスできるサービスとして発行する必要があります。Azure Bot Service を使用すると、ナレッジ ベースを使用してユーザーの質問に回答するボットを作成して発行できます。

1. QnA Maker ページの上部にある 「**Publish (発行)**」 をクリックします。次に、「**Margies Travel KB**」 ページで 「**Publish (発行)**」 をクリックします。
2. サービスがデプロイされたら、「**Create Bot (ボットの作成)**」 をクリックします。ブラウザーの新しいタブで Azure portal が開き、Azure サブスクリプションで Web App Bot を作成できます。
3. Azure portal で、次の設定を使用して Web App Bot を作成します。設定のほとんどは事前に入力されています。
  - **ボット ハンドル**: *ボットの一意の名前*
  - **サブスクリプション**: *使用する Azure サブスクリプション*
  - **リソース グループ**: *使用する QnA Maker リソースを含むリソース グループ*
  - **場所**: *使用する QnA Maker サービスと同じ場所*
  - **価格レベル**: F0
  - **アプリ名**: ***ボット ハンドル**と同じ。自動的に *.azurewebsites.net* が付加されます。
  - **SDK 言語**: *C# または Node.js を選択*
  - **QnA 認証キー**: *QnA ナレッジ ベースの認証キーに自動的に設定されます*
  - **アプリ サービスのプランと場所**: *適切なプランと場所に自動的に設定されます*
  - **アプリケーション インサイト**: オフ
  - **Microsoft アプリ ID とパスワード**: アプリ ID とパスワードを自動作成します。
4. ボットが作成されるのを待ちます。ベルのように見える右上の通知アイコンが、待つ間アニメーション化されます。デプロイの完了が通知されたら、「**リソースに移動**」 をクリックします。または、ホームページで 「**リソース グループ**」 をクリックして Web App Bot を作成したリソース グループを開いてからクリックします。
5. ボットの左側ペインの 「**Settings**」 を探し、「**Test in Web Chat (Web チャットでのテスト)**」 をクリックし、ボットに「**こんにちは、ようこそ!**」というメッセージが表示されるまで待ちます(初期化には数秒かかる場合があります)。
6. テスト チャット インターフェイスを使用して、ボットがナレッジ ベースからの質問に期待どおりに回答することを確認します。たとえば、「*ホテルをキャンセルしたい*」と質問してみます。
